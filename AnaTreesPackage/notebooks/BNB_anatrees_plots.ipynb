```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('/Users/erezcohen/Desktop/uBoone/AnalysisTreesAna/PassedGBDTFiles/multi_BNB_TrainedOn_MCBNB_MCCOSMIC/passedGBDT_BNB_5e19POT_multi_BNB_TrainedOn_MCBNB_MCCOSMIC_allscores_features_and_scores.csv')
data
```

```python
data_maxprotons = data[data.mscore_max==0]
data_maxprotons95 = data_maxprotons[data_maxprotons.mscore_p>0.95]
data_nonmaxprotons = data[data.mscore_max!=0]
data_maxmuon = data[data.mscore_max==1]
```

```python
fig = plt.figure(figsize=[10,10])
bins = np.linspace(0,50,50)
plt.hist(data_maxprotons.pidpida[data_maxprotons.pidpida<100],bins=bins,color='blue',histtype='step',normed=1,label='max scored protons')
plt.hist(data_nonmaxprotons.pidpida[data_nonmaxprotons.pidpida<100],bins=bins,color='green',histtype='step',normed=1,label='max scored non protons')
plt.hist(data_maxmuon.pidpida[data_maxmuon.pidpida<100],bins=bins,color='black',histtype='step',normed=1,label='max scored muons')
plt.hist(data_maxprotons95.pidpida[data_maxprotons95.pidpida<100],bins=bins,color='purple',histtype='step',normed=1,label='max scored protons (score>0.95)')
plt.xlabel('pid-A',fontsize=25)
plt.legend()
```

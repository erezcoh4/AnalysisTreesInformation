```python
ana_MCp = TPlots( anafiles_path + "/Tracks_" + DataName + "_MCprotons_AnalysisTrees.root" , 'TracksTree' )
ana_MCmu = TPlots( anafiles_path + "/Tracks_" + DataName + "_MCmuons_AnalysisTrees.root" , 'TracksTree' )
ana_MCpi = TPlots( anafiles_path + "/Tracks_" + DataName + "_MCpions_AnalysisTrees.root" , 'TracksTree' )
ana_MCem = TPlots( anafiles_path + "/Tracks_" + DataName + "_MCem_AnalysisTrees.root" , 'TracksTree' )
```

```python
branches=['tracks.dEdx_Y','tracks.track_length_Y','tracks.Edep_Y']

def read_tree(ana,label='protons'):
    print 'processing %s data'%label
    ana_arr = tree2array(ana.GetTree(),branches=branches)
    length = np.concatenate(ana_arr['tracks.track_length_Y']).ravel().tolist()
    dEdx = np.concatenate(ana_arr['tracks.dEdx_Y']).ravel().tolist()
    Edep = np.concatenate(ana_arr['tracks.Edep_Y']).ravel().tolist()
    df = pd.DataFrame(data={'residual-range':length,'dE/dx':dEdx,'Edep':Edep})
    print 'read %s data'%label
    return df
```

```python
df_p = read_tree(ana_MCp,label='protons')
df_mu = read_tree(ana_MCmu,label='muon')
df_pi = read_tree(ana_MCpi,label='pions')
df_em = read_tree(ana_MCem,label='em')
```

```python
Nmax = 100000
```

```python
for df , color, label in zip([df_p,df_mu,df_pi,df_em],colors,labels):
    print 'processing plot for ',label
    data = df[df['residual-range']<100]
    data = data[data['residual-range']>0.5]
    data = data[data['dE/dx']>0.5]
    data = data[data['dE/dx']<40]
    print "data['dE/dx'].min():",data['dE/dx'].min()
    print "data['residual-range'].min():",data['residual-range'].min()

    data = data[0:Nmax]
    g=sns.jointplot("residual-range", "dE/dx", data=data,kind="hex", space=0, color=color,
                   marginal_kws=dict(bins=10))
    ax = g.ax_joint
    ax.text(30,15,label,color=color,fontsize=20)
    g.set_axis_labels ("residual range [cm]", "dE/dx [MeV/cm]")
    g.savefig('/Users/erezcohen/Desktop/%s.pdf'%label)
    print 'done generating plot for ',label
```

```python
cols_Float_t=['length','pidpida' , 'pidchi','dqdx_ratio']
def read_tree_fast(filename,label='protons'):
    print 'processing %s data'%label
    f = ROOT.TFile(filename)
    tree = f.Get('TracksTree')
    ana_arr = tree2array(tree,branches=cols_Float_t)
    length = np.concatenate(ana_arr['length']).ravel().tolist()
    pidpida = np.concatenate(ana_arr['pidpida']).ravel().tolist()
    pidchi = np.concatenate(ana_arr['pidchi']).ravel().tolist()
    dqdx_ratio = np.concatenate(ana_arr['dqdx_ratio']).ravel().tolist()
    df = pd.DataFrame(data={'length':length,'pidchi':pidchi,'pidpida':pidpida,'dqdx_ratio':dqdx_ratio})
    df = df[df['dqdx_ratio']<10000]
    print 'read %s data'%label
    return df
```

```python
purity = 0.9999
df_p = read_tree_fast(anafiles_path + "/Tracks_" + DataName + "_MCprotons_trkpurity%f_AnalysisTrees.root"%purity ,label='protons')
df_mu = read_tree_fast(anafiles_path + "/Tracks_" + DataName + "_MCmuons_trkpurity%f_AnalysisTrees.root"%purity,label='muon')
df_pi = read_tree_fast(anafiles_path + "/Tracks_" + DataName + "_MCpions_trkpurity%f_AnalysisTrees.root"%purity,label='pions')
df_em = read_tree_fast(anafiles_path + "/Tracks_" + DataName + "_MCem_trkpurity%f_AnalysisTrees.root"%purity,label='em')
```

```python
def plot_var_x(var,xmin=0,xmax=80,DoAllOnSamePlot=False):
    if DoAllOnSamePlot:
        fig , ax = plt.subplots()
        pt.set_axes(ax , 'tracks %s'%var , '')

    bins = np.linspace(xmin,xmax,100)
    for df , color, label in zip([df_p,df_mu,df_pi,df_em],colors,labels):
        if DoAllOnSamePlot==False:
            fig , ax = plt.subplots()
            pt.set_axes(ax , 'tracks %s'%var , '')
        print 'processing plot for ',label
        plt.hist(df[var], bins=bins, color=color , label=label , alpha=0.3)
        print 'done generating plot for ',label
        plt.legend()
```

```python
plot_var_x('length')
```

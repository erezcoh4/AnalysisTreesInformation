```python
import sys, pandas as pd, numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl , seaborn as sns; sns.set(style="white", color_codes=True , font_scale=1)
%matplotlib inline
sys.path.insert(0, '/Users/erezcohen/larlite/UserDev/mySoftware/MySoftwarePackage/mac')
import GeneralPlot as gp , Initiation as init
dirname = init.createnewdir()
```

```python
hits_file_name = "/Users/erezcohen/Desktop/uBoone/AnalysisTreesAna/TracksAnaFiles/hits_data.csv"
hits = pd.read_csv(hits_file_name)
hits = hits[ hits.dEdx < 60 ] # arbitrary threshold, needs to be studied!
```

```python
plt.figure(figsize=[10,10])
with sns.axes_style("white"):
    g = sns.jointplot(hits.track_length , hits.dEdx , 
                      cmap='hot_r', kind="hex", stat_func=None,
                      xlim=(0, 150), ylim=(0,15),
                      marginal_kws={'color': 'blue'})
    g.set_axis_labels('track length [cm]' , '$dE/dx$ [MeV/cm]')
    plt.colorbar()
g.set_axis_labels('track length [cm]','$dE/dx$ [MeV/cm]')
plt.show()
g.savefig(dirname + "/dEdx_vs_track_length.pdf")
```

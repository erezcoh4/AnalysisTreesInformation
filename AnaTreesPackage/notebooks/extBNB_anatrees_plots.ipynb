```python
fig,ax = plt.subplots(figsize=(14,10))
plt.hist(tracks.length,bins=np.linspace(0,100,1000))
pt.set_axes(ax,'track length [cm]','')
```

```python
fig,ax = plt.subplots(figsize=(14,10))
plt.hist(tracks.cftime,bins=np.linspace(-1,15,100))
pt.set_axes(ax,'closest flash distance [$\mu$s]','')
```

```python
fig,ax = plt.subplots(figsize=(14,10))
plt.hist2d(tracks.startx,tracks.starty,bins=(np.linspace(-1,260,100),np.linspace(-120,120,100)),cmap='winter',norm=LogNorm())
pt.set_axes(ax,'$x_{start}$ [cm]','$y_{start}$ [cm]')
```

```python
fig,ax = plt.subplots(figsize=(14,10))
plt.hist2d(tracks.endx,tracks.endy,bins=(np.linspace(-1,260,100),np.linspace(-120,120,100)),cmap='winter',norm=LogNorm())
pt.set_axes(ax,'$x_{end}$ [cm]','$y_{end}$ [cm]')
```

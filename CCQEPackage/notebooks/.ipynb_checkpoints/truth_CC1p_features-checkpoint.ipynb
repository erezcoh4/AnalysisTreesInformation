```python
import sys; sys.path.insert(0, '../../AnaTreesPackage/notebooks')
from notebooks_tools import * 
%matplotlib inline
```

# Truth CC1p pairs

```python
truthCC1p = pd.read_csv('/Users/erezcohen/Desktop/uBoone/analysis/ccqe_candidates/ccqe_candidates_TruthCC1p_MCC7_MC_BNB.csv',
                     converters={'track_id':generic,'distances':generic,'delta_phi':generic})
print len(truthCC1p),'true cc1p vertices'
# truthCC1p[0:5]
```

```python
GENIECC1p = pd.read_csv('/Users/erezcohen/Desktop/uBoone/analysis/ccqe_candidates/ccqe_candidates_GENIECC1p_MCC7_MC_BNB.csv',
                     converters={'track_id':generic,'distances':generic,'delta_phi':generic})
print len(GENIECC1p),'genie cc1p vertices'
recoGENIECC1p = GENIECC1p[GENIECC1p.IsVertexReconstructed==True]
print len(recoGENIECC1p),'fully reconstructed'
```

# plots to look for enhanecemnt of CC1p search

```python
fig,ax = plt.subplots(figsize=(16,9))
plot_2d_histogram(data=recoGENIECC1p , varx='distance' , vary='delta_phi' ,fontsize=35,
                  x_label = 'rec. tracks distance [cm]', y_label='rec. $\\Delta \\phi$ [deg.]',
                  binsx=np.linspace(0,10,50), binsy=np.linspace(0,360,50), ax=ax , do_save_fig=False,
                  do_add_title=False)
ax.grid(linestyle='--',alpha=0.5)
```

```python
fig,ax = plt.subplots(figsize=(16,9))
plt.plot([0,0],[100,100],linestyle='--',linewidth=3,color='blue')
plot_2d_histogram(data=recoGENIECC1p , varx='PIDA_muon' , vary='PIDA_proton' ,fontsize=35,
                  x_label = 'rec. $PID_A^{\\mu}$', y_label='rec. $PID_A^{p}$',
                  binsx=np.linspace(0,50,50), binsy=np.linspace(0,50,50), ax=ax , norm='non')
ax.grid(linestyle='--',which='both',alpha=0.5)
# zoomin
fig,ax = plt.subplots(figsize=(16,9))
plt.plot([0,0],[100,100],linestyle='--',linewidth=3,color='blue')
plot_2d_histogram(data=recoGENIECC1p , varx='PIDA_muon' , vary='PIDA_proton' ,fontsize=35,
                  x_label = 'rec. $PID_A^{\\mu}$', y_label='rec. $PID_A^{p}$',
                  binsx=np.linspace(0,20,100), binsy=np.linspace(12,25,100), ax=ax , do_save_fig=False,
                  do_add_title=False)
ax.grid(linestyle='--',which='both',alpha=0.5)
```

```python
fig,ax = plt.subplots(figsize=(16,9))
plot_2d_histogram(data=recoGENIECC1p , varx='l_muon' , vary='l_proton' ,fontsize=35,
                  x_label = 'rec. $l_{muon}$ [cm]', y_label='rec. $l_{proton}$ [cm]', 
                  binsx=np.linspace(0,400,100), binsy=np.linspace(0,100,100), ax=ax , do_save_fig=False,
                  do_add_title=False)
ax.grid(linestyle='--',alpha=0.5)
```

```python
fig = plt.figure(figsize=(25,9))
ax = fig.add_subplot(1,2,1)
plot_2d_histogram(data=recoGENIECC1p , varx='l_muon' , vary='truth_P_muon' ,fontsize=35,
                  x_label = 'rec. $l_{\\mu}$ [cm]', y_label='true $p_{\\mu}$ [GeV/c]',
                  binsx=np.linspace(0,300,100), binsy=np.linspace(0,1,100), ax=ax )
ax.grid(linestyle='--',alpha=0.5)

ax = fig.add_subplot(1,2,2)
plot_2d_histogram(data=recoGENIECC1p , varx='l_proton' , vary='truth_P_proton' ,fontsize=35,
                  x_label = 'rec. $l_{p}$ [cm]', y_label='true $p_{p}$ [GeV/c]',
                  binsx=np.linspace(0,100,100), binsy=np.linspace(0,1,100), ax=ax )
ax.grid(linestyle='--',alpha=0.5)
```

```python
fig,ax = plt.subplots(figsize=(16,9))
plot_2d_histogram(data=recoGENIECC1p , varx='delta_theta' , vary='delta_phi' ,fontsize=35,
                  x_label = 'rec.  $\\Delta \\theta$ [deg.]', y_label='rec. $\\Delta \\phi$ [deg.]',
                  binsx=50, binsy=50, ax=ax)
ax.grid(linestyle='--',alpha=0.5)

```

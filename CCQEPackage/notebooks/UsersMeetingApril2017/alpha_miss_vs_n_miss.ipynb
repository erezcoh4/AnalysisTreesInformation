```python
m_n = 0.939
m_mu = 0.1056
```

```python
def generate_uniform_over_a_sphere(r):
    phi = np.random.uniform(0,2*np.pi)
    costheta = np.random.uniform( -1 , 1  )
    theta = np.arccos( costheta )
    x = r * np.sin( theta ) * np.cos( phi )
    y = r * np.sin( theta ) * np.sin( phi )
    t = np.sqrt(np.square(x)+np.square(y))
    z = r * np.cos( theta )
    return x,y,z,t
```

```python
N = 100000
Pt, n_init,n_init_x,n_init_y,n_init_z,n_init_t,E_init , alpha_init = [] , [] , [] , [] , [] , [] ,[] , []
xB, Emu = [], []
for i in range(N):
    if (i%(N/10)==0): print i
    n_init.append(np.random.uniform(0 , 0.75))
    x,y,z,t = generate_uniform_over_a_sphere(n_init[-1])
    n_init_x.append(x)
    n_init_y.append(y)
    n_init_z.append(z)
    n_init_t.append(t)
    E_init.append(np.sqrt(np.square(n_init[-1])+np.square(m_n)))
    alpha_init.append((E_init[-1]-n_init_z[-1])/(0.931))

    # scattering
    Ev = np.random.uniform(0,3)
    nu_x,nu_y,nu_z,nu_t = generate_uniform_over_a_sphere(Ev)
    q = np.random.uniform(0,Ev)
    q_x,q_y,q_z,q_t = generate_uniform_over_a_sphere(q)
    mu_x,mu_y,mu_z = nu_x-q_x , nu_y - q_y , nu_z - q_z
    Emu.append(np.sqrt(np.square(mu_x)+np.square(mu_y)+np.square(mu_z) + np.square(m_mu)))
    omega = Ev - Emu[-1]
    Q2 = np.square(q) - np.square(omega)
    xB.append( Q2/(2*m_n*omega if omega>0.01 else 2*m_n*0.01))
    
    p_x,p_y,p_z = n_init_x[-1] + q_x , n_init_y[-1] + q_y , n_init_z[-1] + q_z

    Pt.append(np.sqrt(np.square(mu_x+p_x) + np.square(mu_y+p_y)))
```

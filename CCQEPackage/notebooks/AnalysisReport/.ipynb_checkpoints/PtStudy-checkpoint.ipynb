## Pt reconstruction for fully contained CC1p events
ask if the resolution effect is smaller than the physical smearing of p_t

```python
GENIECC1p_in_FV_close_proximity['truth_Pt'] = np.sqrt( np.square(GENIECC1p_in_FV_close_proximity['truth_Pp_x']+GENIECC1p_in_FV_close_proximity['truth_Pmu_x'])
                                                     + np.square(GENIECC1p_in_FV_close_proximity['truth_Pp_y']+GENIECC1p_in_FV_close_proximity['truth_Pmu_y']))
plt.hist([GENIECC1p_in_FV_close_proximity['truth_Pt'],GENIECC1p_in_FV_close_proximity['reco_CC1p_Pt']],bins=np.linspace(0,1,10));
```

```python
sample = GENIECC1p_in_FV_close_proximity
fig=plt.figure(figsize=(16,8))

ax = fig.add_subplot(1,2,1)
x = sample['truth_Pt'] - sample['reco_CC1p_Pt']
plt.hist( x , bins=np.linspace(-0.7,0.7,30),color='salmon');
ax.grid(linestyle='--',alpha=0.5)
ax.legend(loc='upper left',fontsize=25)
plt.plot([0,0],ax.get_ylim(),'--',color='black')
set_axes(ax,'$p_{t}^{gen.}-p_{t}^{rec.}$ [GeV/c]',ticks_color='black',fontsize=30,nticklabels=4)
ax.yaxis.set_major_formatter(ticker.NullFormatter())
ax.xaxis.set_ticks([-0.5,0,0.5]);
ax.grid(linestyle='--',alpha=0.9)



ax = fig.add_subplot(1,2,2)
bins = np.linspace(-100,100,30)
ratio = 100*(sample['truth_Pt'] - sample['reco_CC1p_Pt'])/sample['truth_Pt']
plt.hist( ratio , bins=bins,color='royalblue');
ax.grid(linestyle='--',alpha=0.5)
ax.legend(loc='upper left',fontsize=25)
plt.plot([0,0],ax.get_ylim(),'--',color='black')
set_axes(ax,'$(p_{t}^{gen.}-p_{t}^{rec.})/p_{t}^{gen.}$ [%]',ticks_color='black',fontsize=30,nticklabels=4)
ax.xaxis.set_ticks([-80,-40,0,40,80]);
ax.yaxis.set_major_formatter(ticker.NullFormatter())
ax.grid(linestyle='--',alpha=0.9)

plt.tight_layout()
fig.savefig(figures_path+'PtReconstruction.pdf')
```

```python
sample = GENIECC1p_in_FV_close_proximity
fig=plt.figure(figsize=(16,8))

ax = fig.add_subplot(1,2,1)
x = sample['truth_Pt'] - sample['reco_CC1p_Pt']
plot_2d_hist( sample['truth_Pt'] , x , bins=(np.linspace(0,0.5,30),np.linspace(-0.7,0.7,30)),ax=ax,cmap='jet');
ax.grid(linestyle='--',alpha=0.5)
ax.legend(loc='upper left',fontsize=25)
plt.plot([0,0],ax.get_ylim(),'--',color='black')
set_axes(ax,'$p_{t}^{gen.}$ [GeV/c]','$p_{t}^{gen.}-p_{t}^{rec.}$ [GeV/c]',ticks_color='black',fontsize=30,nticklabels=4)
ax.xaxis.set_ticks([0,0.15,0.3,0.45]);
ax.yaxis.set_ticks([-0.5,0,0.5]);
ax.grid(linestyle='--',alpha=0.9)



ax = fig.add_subplot(1,2,2)
ratio = 100*(sample['truth_Pt'] - sample['reco_CC1p_Pt'])/sample['truth_Pt']
plot_2d_hist( sample['truth_Pt'] , ratio , bins=(np.linspace(0,0.5,30),np.linspace(-100,100,30)),ax=ax);
ax.grid(linestyle='--',alpha=0.5)
ax.legend(loc='upper left',fontsize=25)
plt.plot([0,0],ax.get_ylim(),'--',color='black')
set_axes(ax,'$p_{t}^{gen.}$ [GeV/c]','$(p_{t}^{gen.}-p_{t}^{rec.})/p_{t}^{gen.}$ [%]',ticks_color='black',fontsize=30,nticklabels=4)
ax.xaxis.set_ticks([0,0.15,0.3,0.45]);
ax.yaxis.set_ticks([-80,-40,0,40,80]);
ax.grid(linestyle='--',alpha=0.9)

plt.tight_layout()
fig.savefig(figures_path+'PtReconstruction_vs_Pt.pdf')
```
